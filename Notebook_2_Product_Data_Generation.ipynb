```python
pricing_rules = {

    "Electronics": (5000, 80000),

    "Furniture": (2000, 50000),

    "Office Supplies": (20, 2000),

    "Kitchen": (200, 20000),

    "Home Appliances": (3000, 120000),

    "Clothing": (300, 6000),

    "Footwear": (400, 12000),

    "Beauty": (100, 3000),

    "Grocery": (50, 2000),

    "Books": (150, 3000),

    "Sports": (300, 25000),

    "Toys": (200, 8000),

    "Automotive": (300, 30000),

    "Pet Supplies": (200, 5000),

    "Healthcare": (100, 15000)

}
```

```python
stock_rules = {

    "Electronics": (10, 80),

    "Furniture": (5, 40),

    "Office Supplies": (100, 1000),

    "Kitchen": (30, 300),

    "Home Appliances": (10, 60),

    "Clothing": (50, 500),

    "Footwear": (40, 400),

    "Beauty": (100, 800),

    "Grocery": (200, 1500),

    "Books": (50, 500),

    "Sports": (20, 200),

    "Toys": (30, 300),

    "Automotive": (20, 150),

    "Pet Supplies": (20, 200),

    "Healthcare": (50, 500)

}
```

```python
product_ids = [
    f"P{str(i).zfill(5)}"
    for i in range(1, NUMBER_OF_PRODUCTS + 1)
]

print(product_ids[:10])
```

```python
print("=" * 60)

print("PRODUCT DATA GENERATION CONFIGURATION")

print("=" * 60)

print("Total Products :", NUMBER_OF_PRODUCTS)

print("Categories :", len(categories))

print("Suppliers :", len(suppliers))

print("Unique Product IDs :", len(set(product_ids)))

print("=" * 60)
```

```python
product_name_list = []
brand_list = []
category_list = []
sub_category_list = []
supplier_list = []
cost_price_list = []
selling_price_list = []
profit_margin_list = []
stock_quantity_list = []
```

```python
for product_id in product_ids:

    # Select Category
    category = random.choice(list(categories.keys()))

    # Select Sub Category
    sub_category = random.choice(categories[category])

    # Select Brand
    brand = random.choice(brands[category])

    # Supplier
    supplier = random.choice(suppliers)

    # Cost Price
    min_price, max_price = pricing_rules[category]

    cost_price = round(
        random.uniform(min_price, max_price),
        2
    )

    # Selling Price
    markup = random.uniform(1.10, 1.40)

    selling_price = round(
        cost_price * markup,
        2
    )

    # Profit Margin
    profit_margin = round(
        selling_price - cost_price,
        2
    )

    # Stock Quantity
    min_stock, max_stock = stock_rules[category]

    stock_quantity = random.randint(
        min_stock,
        max_stock
    )

    # Product Name
    product_name = f"{brand} {sub_category}"

    # Store Values
    product_name_list.append(product_name)
    brand_list.append(brand)
    category_list.append(category)
    sub_category_list.append(sub_category)
    supplier_list.append(supplier)
    cost_price_list.append(cost_price)
    selling_price_list.append(selling_price)
    profit_margin_list.append(profit_margin)
    stock_quantity_list.append(stock_quantity)
```

```python
products_df = pd.DataFrame({

    "product_id": product_ids,

    "product_name": product_name_list,

    "brand": brand_list,

    "category": category_list,

    "sub_category": sub_category_list,

    "supplier": supplier_list,

    "cost_price": cost_price_list,

    "selling_price": selling_price_list,

    "profit_margin": profit_margin_list,

    "stock_quantity": stock_quantity_list

})
```
