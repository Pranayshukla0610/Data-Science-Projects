```python
import random
import pandas as pd
import numpy as np
```

```python
random.seed(42)
np.random.seed(42)
```

```python
orders_df = pd.read_csv('/content/orders (2).csv')
```

```python
products_df = pd.read_csv('/content/products (1).csv')
```

```python
print("Orders :", orders_df.shape)
print("Products :", products_df.shape)
```

```python
product_price_lookup = dict(
    zip(
        products_df["product_id"],
        products_df["selling_price"]
    )
)

product_profit_lookup = dict(
    zip(
        products_df["product_id"],
        products_df["profit_margin"]
    )
)
```

```python
product_ids = products_df['product_id'].tolist()
```

```python
order_ids = orders_df['order_id'].tolist()
```

```python
MIN_PRODUCTS_PER_ORDER = 1
MAX_PRODUCTS_PER_ORDER = 5
```

```python
discount_options = [
    0,
    5,
    10,
    15,
    20,
    25,
    30
]
```

```python
discount_probability = [
    0.30,
    0.20,
    0.20,
    0.10,
    0.10,
    0.05,
    0.05
]
```

```python
quantity_options = [
    1,
    2,
    3,
    4,
    5
]
```

```python
quantity_probability = [
    0.45,
    0.25,
    0.15,
    0.10,
    0.05
]
```

```python
order_detail_id_list = []

order_id_list = []

product_id_list = []

quantity_list = []

unit_price_list = []

discount_list = []

line_sales_list = []

line_profit_list = []
```

```python
detail_counter = 1
```

```python
print("=" * 60)

print("ORDER DETAILS CONFIGURATION")

print("=" * 60)

print("Orders :", len(order_ids))

print("Products :", len(product_ids))

print("Products per Order :",
      f"{MIN_PRODUCTS_PER_ORDER}-{MAX_PRODUCTS_PER_ORDER}")

print("=" * 60)
```

```python
estimated_rows = len(order_ids) * 3

print("Estimated Order Detail Records:", estimated_rows)
```

```python
for order_id in order_ids:

    # Number of products in this order
    number_of_products = random.randint(
        MIN_PRODUCTS_PER_ORDER,
        MAX_PRODUCTS_PER_ORDER
    )

    # Ensure unique products in an order
    selected_products = random.sample(
        product_ids,
        number_of_products
    )

    for product in selected_products:

        # Quantity
        quantity = np.random.choice(
            quantity_options,
            p=quantity_probability
        )

        # Unit Price
        unit_price = product_price_lookup[product]

        # Discount
        discount = np.random.choice(
            discount_options,
            p=discount_probability
        )

        # Gross Sales
        gross_sales = quantity * unit_price

        # Discount Amount
        discount_amount = gross_sales * (discount / 100)

        # Final Sales
        line_sales = round(
            gross_sales - discount_amount,
            2
        )

        # Profit Margin
        profit_margin = product_profit_lookup[product]

        # Line Profit
        line_profit = round(
            (profit_margin * quantity) - discount_amount,
            2
        )

        # Order Detail ID
        order_detail_id = f"OD{detail_counter:08d}"

        detail_counter += 1

        # Store Values
        order_detail_id_list.append(order_detail_id)

        order_id_list.append(order_id)

        product_id_list.append(product)

        quantity_list.append(quantity)

        unit_price_list.append(unit_price)

        discount_list.append(discount)

        line_sales_list.append(line_sales)

        line_profit_list.append(line_profit)
```

```python
order_details_df = pd.DataFrame({

    "order_detail_id": order_detail_id_list,

    "order_id": order_id_list,

    "product_id": product_id_list,

    "quantity": quantity_list,

    "unit_price": unit_price_list,

    "discount_percent": discount_list,

    "line_sales": line_sales_list,

    "line_profit": line_profit_list

})
```
