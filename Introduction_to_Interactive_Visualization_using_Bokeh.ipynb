```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

X = [1,2,3,4,5]
y = [10,20,30,40,50]

p = figure(
    title='Simple Line Chart',
    x_axis_label = 'X Axis',
    y_axis_label = 'Y Axis',
    width = 700,
    height = 400
)

p.line(X, y, line_width=3)
show(p)
```

```python
X = [1,2,3,4,5]
y = [5,7,2,8,4]

p = figure(title='Scatter Plot')

p.circle(
    X,
    y,
    size=15,
    color='navy',
    alpha=0.5
)

show(p)
```

```python
from bokeh.models import HoverTool

hover = HoverTool(
    tooltips = [
        ("index","$index"),
        ('Value','@y')
    ]
)

p.add_tools(hover)
show(p)
```

```python
from bokeh.models import ColumnDataSource
import pandas as pd

df = pd.DataFrame({
    'x':[1,2,3,4],
    'y':[10,20,15,30]
})

source = ColumnDataSource(df)

p = figure()

p.line(
    x='x',
    y='y',
    source = source,
    line_width=3
)

show(p)
```

```python
import yfinance as yf
from bokeh.plotting import figure, show

data = yf.download("AAPL", start='2014-01-01')

p = figure(
    x_axis_type='datetime',
    title='Apple Stock Price',
    width = 900,
    height=400
)

p.line(
    data.index,
    data['Close'],
    line_width=2
)

show(p)
```

```python
import pandas as pd
import yfinance as yf

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Download stock data
df = yf.download("AAPL", period="3mo", auto_adjust=False)

# Remove MultiIndex columns
df.columns = df.columns.get_level_values(0)

# Reset index
df.reset_index(inplace=True)

# Boolean conditions
inc = df["Close"] > df["Open"]
dec = df["Open"] > df["Close"]

# Candle width
w = 12 * 60 * 60 * 1000

# Create figure
p = figure(
    x_axis_type="datetime",
    width=1000,
    height=400,
    title="AAPL Candlestick Chart"
)

# High-Low line
p.segment(
    df["Date"],
    df["High"],
    df["Date"],
    df["Low"],
    color="black"
)

# Green candles
p.vbar(
    x=df.loc[inc, "Date"],
    width=w,
    top=df.loc[inc, "Close"],
    bottom=df.loc[inc, "Open"],
    fill_color="green",
    line_color="black"
)

# Red candles
p.vbar(
    x=df.loc[dec, "Date"],
    width=w,
    top=df.loc[dec, "Open"],
    bottom=df.loc[dec, "Close"],
    fill_color="red",
    line_color="black"
)

show(p)
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
import numpy as np

# Enable Bokeh in notebook
output_notebook()

# Data
X = np.arange(10)
y = np.random.rand(10)

# Data source
source = ColumnDataSource(data=dict(X=X, y=y))

# Figure
p = figure(
    width=700,
    height=400,
    title="Random Line Chart"
)

# Line
p.line('X', 'y', source=source, line_width=3)

# Show graph
show(p)
```
