```python
import random
import pandas as pd
import numpy as np

from faker import Faker
from datetime import datetime
```

```python
fake = Faker("en_IN")
```

```python
random.seed(42)
np.random.seed(42)
Faker.seed(42)
```

```python
NUMBER_OF_CUSTOMERS = 15000
```

```python
segments = [
    'Consumer',
    'Corporate',
    'Home Office'
]
```

```python
segment_probability = [
    0.60,
    0.25,
    0.15
]
```

```python
genders = [
    'Male',
    'Female'
]
```

```python
gender_probability = [
    0.50,
    0.50
]
```

```python
city_state = {

    "Lucknow":"Uttar Pradesh",
    "Kanpur":"Uttar Pradesh",
    "Varanasi":"Uttar Pradesh",
    "Agra":"Uttar Pradesh",
    "Noida":"Uttar Pradesh",

    "Delhi":"Delhi",

    "Mumbai":"Maharashtra",
    "Pune":"Maharashtra",
    "Nagpur":"Maharashtra",
    "Nashik":"Maharashtra",

    "Bengaluru":"Karnataka",
    "Mysuru":"Karnataka",
    "Hubli":"Karnataka",

    "Hyderabad":"Telangana",
    "Warangal":"Telangana",

    "Chennai":"Tamil Nadu",
    "Coimbatore":"Tamil Nadu",
    "Madurai":"Tamil Nadu",

    "Jaipur":"Rajasthan",
    "Jodhpur":"Rajasthan",

    "Ahmedabad":"Gujarat",
    "Surat":"Gujarat",
    "Vadodara":"Gujarat",

    "Bhopal":"Madhya Pradesh",
    "Indore":"Madhya Pradesh",

    "Patna":"Bihar",
    "Gaya":"Bihar",

    "Kolkata":"West Bengal",
    "Siliguri":"West Bengal",

    "Bhubaneswar":"Odisha",
    "Cuttack":"Odisha",

    "Chandigarh":"Punjab",

    "Amritsar":"Punjab",

    "Kochi":"Kerala",
    "Thiruvananthapuram":"Kerala",

    "Guwahati":"Assam",

    "Ranchi":"Jharkhand",

    "Raipur":"Chhattisgarh",

    "Dehradun":"Uttarakhand",

    "Shimla":"Himachal Pradesh"
}
```

```python
cities = list(city_state.keys())
```

```python
MIN_AGE = 18
MAX_AGE = 70
```

```python
print("Total Customers :", NUMBER_OF_CUSTOMERS)

print("Cities :", len(cities))

print("Segments :", segments)

print("Gender :", genders)

print("Minimum Age :", MIN_AGE)

print("Maximum Age :", MAX_AGE)
```

```python
customer_ids = [
    f"C{str(i).zfill(5)}"
    for i in range(1, NUMBER_OF_CUSTOMERS + 1)
]
print(customer_ids[:10])
```

```python
print("Number of Customer IDs Generated:", len(customer_ids))

print("Unique Customer IDs:", len(set(customer_ids)))
```

```python
customer_name_list = []

gender_list = []

age_list = []

city_list = []

state_list = []

segment_list = []
```

```python
for customer in customer_ids:
  customer_name = fake.name()
  gender = np.random.choice(
      genders,
      p=gender_probability
  )

  age = random.randint(
      MIN_AGE,
      MAX_AGE
  )

  city = random.choice(cities)

  state = city_state[city]

  segment = np.random.choice(
      segments,
      p=segment_probability
  )

  customer_name_list.append(customer_name)
  gender_list.append(gender)
  age_list.append(age)
  city_list.append(city)
  state_list.append(state)
  segment_list.append(segment)
```

```python
customers_df = pd.DataFrame({
    "customer_id":customer_ids,
    "customer_name":customer_name_list,
    "gender":gender_list,
    "age":age_list,
    "city":city_list,
    "state":state_list,
    "segment":segment_list
})
```
