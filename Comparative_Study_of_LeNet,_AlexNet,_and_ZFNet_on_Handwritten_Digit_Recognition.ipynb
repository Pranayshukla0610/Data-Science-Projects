```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import pandas as pd
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape)
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
X_train = np.expand_dims(X_train,-1)
X_test = np.expand_dims(X_test,-1)
```

```python
print(X_train.shape)
```

```python
#ONE HOT ENCODING
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
#LeNet Layers

from tensorflow.keras import models, layers

LeNet = models.Sequential()

LeNet.add(
    layers.Conv2D(
        filters=6,
        kernel_size=(5,5),
        activation='tanh',
        input_shape=(28,28,1)
    )
)

LeNet.add(
    layers.AveragePooling2D(
        pool_size=(2,2)
    )
)

LeNet.add(
    layers.Conv2D(
        filters=16,
        kernel_size=(5,5),
        activation='tanh'
    )
)

LeNet.add(
    layers.AveragePooling2D(
        pool_size=(2,2)
    )
)

LeNet.add(layers.Flatten())

LeNet.add(
    layers.Dense(
        120,
        activation='tanh'
    )
)

LeNet.add(
    layers.Dense(
        84,
        activation='tanh'
    )
)

LeNet.add(
    layers.Dense(
        10,
        activation='softmax'
    )
)

LeNet.summary()
```

```python
LeNet.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_lenet = LeNet.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.2
)
```

#AlexNet

```python
#Resize Images
#AlexNet expects: 227x227x3, it is too large for CPU, so use 64x64x1

X_train_64 = tf.image.resize(
    X_train,
    (64,64)
)

X_test_64 = tf.image.resize(
    X_test,
    (64,64)
)
```

```python
#AlexNet Architrecture

#Conv 96
#↓

#Max Pool

#↓

#Conv 256

#↓

#Max Pool

#↓

#Conv 384

#↓

#Conv 384

#↓

#Conv 256

#↓

#Max Pool

#↓

#FC 4096

#↓

#FC 4096

#↓

#Output
```

```python
from tensorflow.keras import models,layers

AlexNet = models.Sequential([

    layers.Input(shape=(64,64,1)),


    layers.Conv2D(
        32,
        (11,11),
        strides=2,
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(3,3),
        strides=2
    ),


    layers.Conv2D(
        64,
        (5,5),
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(3,3),
        strides=2
    ),


    layers.Conv2D(
        128,
        (3,3),
        padding='same',
        activation='relu'
    ),


    layers.Conv2D(
        128,
        (3,3),
        padding='same',
        activation='relu'
    ),


    layers.Conv2D(
        64,
        (3,3),
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(2,2)
    ),

    layers.Flatten(),

    layers.Dense(
        512,
        activation='relu'
    ),

    layers.Dropout(0.5),

    layers.Dense(
        256,
        activation='relu'
    ),

    layers.Dropout(0.5),

    layers.Dense(
        10,
        activation='softmax'
    )
])

AlexNet.summary()
```

```python
AlexNet.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_alex=AlexNet.fit(
    X_train_64,
    y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.2
)
```

```python
ZFNet = models.Sequential([

    layers.Input(shape=(64,64,1)),

    layers.Conv2D(
        32,
        (7,7),
        strides=2,
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(3,3),
        strides=2
    ),

    layers.Conv2D(
        64,
        (5,5),
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(3,3),
        strides=2
    ),

    layers.Conv2D(
        128,
        (3,3),
        padding='same',
        activation='relu'
    ),

    layers.Conv2D(
        128,
        (3,3),
        padding='same',
        activation='relu'
    ),

    layers.Conv2D(
        64,
        (3,3),
        padding='same',
        activation='relu'
    ),

    layers.MaxPooling2D(
        pool_size=(2,2)
    ),

    layers.Flatten(),

    layers.Dense(
        512,
        activation='relu'
    ),

    layers.Dropout(0.5),

    layers.Dense(
        256,
        activation='relu'
    ),

    layers.Dropout(0.5),

    layers.Dense(
        10,
        activation='softmax'
    )
])

ZFNet.summary()
```

```python
ZFNet.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = LeNet.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    callbacks=[early_stop]
)
```

```python
lenet_acc=LeNet.evaluate(
    X_test,
    y_test,
    verbose=0
)

alex_acc=AlexNet.evaluate(
    X_test_64,
    y_test,
    verbose=0
)

zf_acc=ZFNet.evaluate(
    X_test_64,
    y_test,
    verbose=0
)

print(lenet_acc)
print(alex_acc)
print(zf_acc)
```
