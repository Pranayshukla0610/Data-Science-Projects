```python
import random
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
```

```python
random.seed(42)
np.random.seed(42)
```

```python
NUMBER_OF_ORDERS = 100000
```

```python
customer_df = pd.read_csv('/content/customers (1).csv')
```

```python
products_df = pd.read_csv('/content/products (1).csv')
```

```python
regions_df = pd.read_csv('/content/regions.csv')
```

```python
print(customer_df.shape)
print(products_df.shape)
print(regions_df.shape)
```

```python
customer_ids = customer_df['customer_id'].tolist()
```

```python
product_ids = products_df['product_id'].tolist()
```

```python
region_ids = regions_df['region_id'].tolist()
```

```python
print(len(customer_ids))
print(len(product_ids))
print(region_ids)
```

```python
order_ids = [
    f"O{str(i).zfill(7)}"
    for i in range(
        1,
        NUMBER_OF_ORDERS+1
    )
]
```

```python
order_ids[:10]
```

```python
payment_methods = [
    'UPI',
    'Credit Card',
    'Debit Card',
    'Net Banking',
    'Cash on Delivery',
    'Wallet'
]
```

```python
payment_probability = [
    0.40,
    0.20,
    0.15,
    0.05,
    0.15,
    0.05
]
```

```python
order_status = [
    'Delivered',
    'Shipped',
    'Cancelled',
    'Returned'
]
```

```python
status_probability = [
    0.88,
    0.07,
    0.03,
    0.02
]
```

```python
discount_options = [
    0,
    5,
    10,
    15,
    20,
    25,
    30
]
```

```python
discount_probability = [

    0.30,

    0.20,

    0.20,

    0.10,

    0.10,

    0.05,

    0.05

]
```

```python
quantity_options = [
    1,
    2,
    3,
    4,
    5
]
```

```python
quantity_probability = [
    0.45,
    0.25,
    0.15,
    0.10,
    0.05
]
```

```python
start_date = datetime(2023,1,1)
end_date = datetime(2026,12,31)
```

```python
date_difference = (
    end_date - start_date
).days
```

```python
print("="*60)

print("ORDER DATA CONFIGURATION")

print("="*60)

print("Orders :",NUMBER_OF_ORDERS)

print("Customers :",len(customer_ids))

print("Products :",len(product_ids))

print("Regions :",len(region_ids))

print("Payment Methods :",payment_methods)

print("Order Status :",order_status)

print("="*60)
```

```python
product_price_lookup = dict(
    zip(
        products_df["product_id"],
        products_df["selling_price"]
    )
)
```

```python
list(product_price_lookup.items())[:5]
```

```python
product_profit_lookup = dict(
    zip(
        products_df["product_id"],
        products_df["profit_margin"]
    )
)
```

```python
print("Everything Loaded Successfully!")

print()

print("Ready to Generate 100,000 Orders")
```

```python
order_date_list = []
ship_date_list = []

customer_id_list = []
product_id_list = []
region_id_list = []

quantity_list = []

payment_method_list = []
order_status_list = []

discount_list = []

unit_price_list = []

sales_amount_list = []

profit_list = []

shipping_cost_list = []
```

```python
customer_region_lookup = {}

for customer in customer_ids:
    customer_region_lookup[customer] = random.choice(region_ids)
```

```python
for order_id in order_ids:

    # -------------------------
    # Customer
    # -------------------------
    customer = random.choice(customer_ids)

    # Region (fixed for customer)
    region = customer_region_lookup[customer]

    # Product
    product = random.choice(product_ids)

    # Quantity
    quantity = np.random.choice(
        quantity_options,
        p=quantity_probability
    )

    # Unit Price
    unit_price = product_price_lookup[product]

    # Discount
    discount = np.random.choice(
        discount_options,
        p=discount_probability
    )

    # Gross Sales
    gross_sales = unit_price * quantity

    # Discount Amount
    discount_amount = gross_sales * (discount / 100)

    # Final Sales
    sales_amount = round(
        gross_sales - discount_amount,
        2
    )

    # Profit
    unit_profit = product_profit_lookup[product]

    profit = round(
        (unit_profit * quantity) - discount_amount,
        2
    )

    # Shipping Cost
    shipping_cost = round(
        random.uniform(50, 500),
        2
    )

    # Payment Method
    payment = np.random.choice(
        payment_methods,
        p=payment_probability
    )

    # Order Status
    status = np.random.choice(
        order_status,
        p=status_probability
    )

    # Random Order Date
    order_date = start_date + timedelta(
        days=random.randint(
            0,
            date_difference
        )
    )

    # Shipping Date
    ship_date = order_date + timedelta(
        days=random.randint(1, 7)
    )

    # Store Values
    customer_id_list.append(customer)

    product_id_list.append(product)

    region_id_list.append(region)

    quantity_list.append(quantity)

    payment_method_list.append(payment)

    order_status_list.append(status)

    discount_list.append(discount)

    unit_price_list.append(unit_price)

    sales_amount_list.append(sales_amount)

    profit_list.append(profit)

    shipping_cost_list.append(shipping_cost)

    order_date_list.append(order_date)

    ship_date_list.append(ship_date)
```

```python
print("order_ids               :", len(order_ids))
print("customer_id_list        :", len(customer_id_list))
print("product_id_list         :", len(product_id_list))
print("region_id_list          :", len(region_id_list))
print("order_date_list         :", len(order_date_list))
print("ship_date_list          :", len(ship_date_list))
print("quantity_list           :", len(quantity_list))
print("payment_method_list     :", len(payment_method_list))
print("order_status_list       :", len(order_status_list))
print("discount_list           :", len(discount_list))
print("unit_price_list         :", len(unit_price_list))
print("sales_amount_list       :", len(sales_amount_list))
print("profit_list             :", len(profit_list))
print("shipping_cost_list      :", len(shipping_cost_list))
```

```python
orders_df = pd.DataFrame({

    "order_id": order_ids,

    "customer_id": customer_id_list,

    "product_id": product_id_list,

    "region_id": region_id_list,

    "order_date": order_date_list,

    "ship_date": ship_date_list,

    "quantity": quantity_list,

    "payment_method": payment_method_list,

    "order_status": order_status_list,

    "discount_percent": discount_list,

    "unit_price": unit_price_list,

    "sales_amount": sales_amount_list,

    "profit": profit_list,

    "shipping_cost": shipping_cost_list

})
```
