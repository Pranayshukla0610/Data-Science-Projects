```python
import random
import pandas as pd
import numpy as np

from datetime import timedelta
from datetime import datetime
```

```python
random.seed(42)
np.random.seed(42)
```

```python
products_df = pd.read_csv('/content/products (1).csv')
```

```python
regions_df = pd.read_csv('/content/regions.csv')
```

```python
print("Products :", products_df.shape)
print("Regions :", regions_df.shape)
```

```python
product_ids = products_df["product_id"].tolist()
```

```python
region_ids = regions_df["region_id"].tolist()
```

```python
product_price_lookup = dict(
    zip(
        products_df["product_id"],
        products_df["selling_price"]
    )
)
```

```python
warehouse_names = [
    'North Warehouse',
    'South Warehouse',
    'East Warehouse',
    'West Warehouse',
    'Central Warehouse'
]
```

```python
warehouse_lookup = {}

for _,row in regions_df.iterrows():
  zone = row['zone']

  if zone == 'North':
    warehouse = 'North Warehouse'
  elif zone == 'South':
    warehouse = 'South Warehouse'
  elif zone == 'West':
    warehouse = 'West Warehouse'
  elif zone == 'East':
    warehouse = 'East Warehouse'
  else:
    warehouse = 'Central Warehouse'

  warehouse_lookup[row["region_id"]] = warehouse
```

```python
start_date = datetime(2023, 1, 1)
end_date = datetime(2026, 12, 31)

date_difference = (
    end_date - start_date
).days
```

```python
NUMBER_OF_PRODUCTS = len(product_ids)

NUMBER_OF_REGIONS = len(region_ids)

print("Products :", NUMBER_OF_PRODUCTS)

print("Regions :", NUMBER_OF_REGIONS)
```

```python
inventory_id_list = []

product_id_list = []

region_id_list = []

warehouse_list = []

opening_stock_list = []

stock_received_list = []

stock_sold_list = []

current_stock_list = []

reorder_level_list = []

stock_status_list = []

inventory_value_list = []

last_restock_date_list = []
```

```python
stock_status_options = [

    "Low Stock",

    "In Stock",

    "Overstock"

]
```

```python
print("=" * 60)

print("INVENTORY CONFIGURATION")

print("=" * 60)

print("Products :", len(product_ids))

print("Regions :", len(region_ids))

print("Expected Inventory Records :",
      len(product_ids) * len(region_ids))

print("=" * 60)
```

```python
inventory_counter = 1
```

```python
for region in region_ids:

    warehouse = warehouse_lookup[region]

    for product in product_ids:

        # Inventory ID
        inventory_id = f"INV{inventory_counter:06d}"

        inventory_counter += 1

        # Opening Stock
        opening_stock = random.randint(100, 1000)

        # Stock Received
        stock_received = random.randint(50, 500)

        # Stock Sold
        max_stock_available = opening_stock + stock_received

        stock_sold = random.randint(
            0,
            max_stock_available
        )

        # Current Stock
        current_stock = (
            opening_stock
            + stock_received
            - stock_sold
        )

        # Reorder Level
        reorder_level = random.randint(80, 250)

        # Stock Status
        if current_stock <= reorder_level:

            stock_status = "Low Stock"

        elif current_stock >= reorder_level * 3:

            stock_status = "Overstock"

        else:

            stock_status = "In Stock"

        # Selling Price
        selling_price = product_price_lookup[product]

        # Inventory Value
        inventory_value = round(
            current_stock * selling_price,
            2
        )

        # Last Restock Date
        last_restock_date = (
            start_date
            + timedelta(
                days=random.randint(
                    0,
                    date_difference
                )
            )
        )

        # Store Values

        inventory_id_list.append(inventory_id)

        product_id_list.append(product)

        region_id_list.append(region)

        warehouse_list.append(warehouse)

        opening_stock_list.append(opening_stock)

        stock_received_list.append(stock_received)

        stock_sold_list.append(stock_sold)

        current_stock_list.append(current_stock)

        reorder_level_list.append(reorder_level)

        stock_status_list.append(stock_status)

        inventory_value_list.append(inventory_value)

        last_restock_date_list.append(last_restock_date)
```

```python
inventory_df = pd.DataFrame({

    "inventory_id": inventory_id_list,

    "product_id": product_id_list,

    "region_id": region_id_list,

    "warehouse_name": warehouse_list,

    "opening_stock": opening_stock_list,

    "stock_received": stock_received_list,

    "stock_sold": stock_sold_list,

    "current_stock": current_stock_list,

    "reorder_level": reorder_level_list,

    "stock_status": stock_status_list,

    "inventory_value": inventory_value_list,

    "last_restock_date": last_restock_date_list

})
```
