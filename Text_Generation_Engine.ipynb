```python
from transformers import pipeline

# Load model
generator = pipeline(
    "text-generation",
    model="gpt2"
)

print("=" * 50)
print("WELCOME TO AI TEXT GENERATOR")
print("=" * 50)

while True:

    prompt = input("\nEnter Prompt: ")

    if prompt.lower() == "exit":
        print("\nExiting Program...")
        break

    try:

        results = generator(
            prompt,
            max_length=100,
            num_return_sequences=2,
            do_sample=True,
            temperature=0.8,
            top_k=50,
            top_p=0.95
        )

        for index, output in enumerate(results):

            print(f"\nGenerated Text {index+1}")
            print("-" * 40)

            generated_text = output['generated_text']

            print(generated_text)

            with open("outputs.txt", "a") as file:
                file.write(generated_text)
                file.write("\n\n")

    except Exception as e:

        print("Error:", e)
```
